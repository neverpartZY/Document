```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.signal import savgol_filter
from matplotlib import font_manager
my_font=font_manager.FontProperties(fname='C:\Windows\Fonts\simhei.ttf')
```

```python
data = pd.read_csv(r'D:\实验文件\仪器分析实验\几种半导体的光谱分析实验数据\12周化学XRD数据\12周周二晚\TiO2.ASC',sep='      ',names=['2theta','indensity'],header=0,)
data
```

```python
x = data['2theta']
y = data['indensity']
y1 = savgol_filter(y, 55, 8)
```

```python
plt.plot(x,y1,label='$TiO_2$',c='k')
a = [i for i in range(5,4000,500)]
plt.xticks(a)
plt.yticks([])
#plt.xlim(5,70)
plt.xlabel(r'$2\theta$',fontsize=14)
plt.ylabel('$intensity(a.u.)$',fontsize=14)
ax = plt.gca()
ax.spines['left'].set_linewidth(3)
ax.spines['right'].set_linewidth(3)
ax.spines['top'].set_linewidth(3)
ax.spines['bottom'].set_linewidth(3)
plt.legend(frameon=False)
plt.show()

```
