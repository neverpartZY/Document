# 绘制3y轴图

```python
f = pd.read_excel('D:\实验文件\化工原理实验\离心泵特征曲线.xlsx',header=1)
f1 = f.iloc[:15,8:13].drop(11,axis=0)

f1.head()
```

```python
plt.style.use('seaborn-white')
x = f1['流量/(m3/s）']
y1 = f1['扬程（m）']
y2 = f1['电功率（w）']
y3 = f1['效率（%）']

fig, ax1 =plt.subplots()
ax2 = ax1.twinx() # twin for second y-axis
ax3 = ax1.twinx() # twin for second y-axis
ax3.spines["right"].set_position(("axes",1.2)) # position of right spine
matplotlib.rcParams['font.family']=['SimHei'] #用中文表述
ax1.plot(x, y1,'ob', markerfacecolor='white',label='扬程')
ax1.set_ylabel('扬程', fontname='SimHei', fontsize=13)
ax1.legend()
ax1.set_ylim(0,35) #x轴和y轴范围 （0,1）*(0,150)不要改


matplotlib.rcParams['font.family']=['SimHei'] #用中文表述
ax2.plot(x, y2,'or', markerfacecolor='white',label='电功率')
ax2.set_ylabel('电功率', fontname='SimHei', fontsize=13)
ax2.legend()


matplotlib.rcParams['font.family']=['SimHei'] #用中文表述
ax3.plot(x, y3,'og', markerfacecolor='white',label='效率')
ax3.set_ylabel('效率', fontname='SimHei', fontsize=13)
ax3.legend()
```
