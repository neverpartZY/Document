```python
import pandas as pd
import numpy as np 		
import matplotlib.pyplot as plt

f = pd.read_excel("D:\实验及上课文件\化工原理实验\传热.xlsx",header=1)
f.iloc[:,-2:].head()
```

```python
f1 = f.iloc[:15,-2:]
x = f1['ln(Re)']
y = f1['ln(Nu/Pr)']
f1
```

```python
from scipy.stats import pearsonr

r , p = pearsonr(x,y)
print(r)
p
a = np.polyfit(x,y,1)
y2 = np.poly1d(a) #y = a0 + a1*x 的函数形式
a
```

```python
aig={
    'font.family':'SimHei',
    'axes.unicode_minus':'False',
    'mathtext.fontset':'stix',
}
plt.rcParams.update(aig)  # 将matplotlib的参数变量替换为你自己的参数。这是
```

```python
fig, ax = plt.subplots(dpi=300)
ax.plot(x,y,'o',markerfacecolor='white',label='套管换热器')
ax.set_xlabel(r'$\ln(Re)$',fontsize=15) 	#设置x轴文本
ax.set_ylabel(r'$\ln\left(\frac{Nu}{Pr^{0.4}}\right)$',fontsize=15) 	#设置y轴文本
ax.set_xlim(9.25,10.8)
ax.set_ylim(-2.4,-0.75)

ax.plot(x,y2(x),'--',label='拟合曲线') 	#根据平面套管
ax.text(9.6,-1,f'$y=${a[0]:.2f}$x${a[1]:.2f}\n$R=${r:.3f}',fontsize=13)

ax.legend(frameon=False)
```
