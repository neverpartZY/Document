```python

from mpl_toolkits.axes_grid1 import host_subplot
from mpl_toolkits import axisartist
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import  font_manager
plt.rcParams['font.family']=['SimHei']#设置字体为harmony 
plt.rcParams['axes.unicode_minus']=False#设置正常显示
```

```python

f = pd.read_excel('D:\实验及上课文件\化工原理实验\离心泵特征曲线.xlsx',header=1)
f1 = f.iloc[:15,8:13].drop(11,axis=0)
f2 = f.iloc[:20,[16,17,20]]
f2.head()
```

```python
plt.figure(dpi=300)
host = host_subplot(111,axes_class=axisartist.Axes,)
plt.subplots_adjust(right=1)

par1 = host.twinx()
par2 = host.twinx()

par2.axis["right"] = par2.new_fixed_axis(loc="right", offset=(60, 0))

par1.axis["right"].toggle(all=True)
par2.axis["right"].toggle(all=True)

x = f1['流量/(m3/s）']
y1 = f1['电功率（w）']
y2 = f1['扬程（m）']
y3 = f1['效率（%）']
p1, = host.plot(x,y1, 'o', label="功率",markersize=8)
p2, = par1.plot(x,y2, 'x', label="扬程",markersize=8)
p3, = par2.plot(x, y3, '^', label="效率",markersize=8)
host.set_ylim(400,1000)
""" host.set_xlim(0, 12)   # 坐标轴长度"""

par1.set_ylim(0, 35)
par2.set_ylim(40, 100) 
host.set_title('离心泵特征曲线',fontsize=16)

host.set_xlabel(r"流量/($m^3/s$)",fontsize=20)
host.set_ylabel("功率/W",  fontsize=20)
par1.set_ylabel("扬程/m",fontsize=20)
par2.set_ylabel("效率/%",fontsize=20)

host.legend(frameon = False, fontsize=10)

host.axis["left"].label.set_color(p1.get_color())
par1.axis["right"].label.set_color(p2.get_color())
par2.axis["right"].label.set_color(p3.get_color())
par1.axis['right']. major_ticklabels.set_color(p2.get_color())
plt.show()

```
