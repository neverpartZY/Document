```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import physical_constants
```

```python
n = 0.5
Eg = 1.4
h = physical_constants["Planck constant"][0]
e = physical_constants['elementary charge'][0]
v = np.arange(200,800)*1e-9
alpha_h_v = h*v/e*np.sin(v)
y =alpha_h_v**n  
B = 1                 
x= B*(h*v/e-Eg)
plt.plot(x,y)                                                         
```

```python
x = np.linspace(0,840,20000)

def y2(x):
    if 0<x<=80:
        y = 5 * np.abs(x-40)
        return y
    else:
        a = x%80
        y = 5 * np.abs(a-40)
        return y
def y1(x):
    if x<=200/3:
        y = -6 * np.abs(x - 100/3) +200
        return y
    else:
        a = x % (200/3)
        y = -6 * np.abs(a - 100/3) +200
        return y
y1 = list(map(y1,x))
y2 = list(map(y2,x))   
fig,ax=plt.subplots(figsize=(16,8),dpi=300)
plt.plot(x,y2,label='T=80')
plt.plot(x,y1,label=r'T=$\frac{200}{3}$')
ax.set_ylim(0,230)
ax.set_xlim(0,840)
plt.legend()
```
