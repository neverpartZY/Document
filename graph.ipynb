```python
import numpy as np
import matplotlib.pyplot as plt
import fractions
def ticks_dyi(x):
    a = round(x/np.pi,2)
    if a == 0:
        return '0'
    elif a**2 == 1:
        return '$\pi$'
    elif a !=int(a):
        b = round(fractions.Fraction(a).numerator)
        c = round(fractions.Fraction(a).denominator)
        return r'$\frac{{%d}}{{%d}}\pi$'%(b,c)
        # return r'$\frac{{}}{{}}\pi$'.format(b,c)

    elif a == int(a) and a!=0:
        return rf'${round(a)}\pi$'
a = [i for i in np.arange(-2*np.pi,2*np.pi+0.1,np.pi/4)]
r = list(map(ticks_dyi,a))
```

```python
x = np.linspace(-2*np.pi,np.pi*2,1000)
y = np.cos(x)
y2 = np.sin(x)
fig, ax=plt.subplots(figsize=(10,6),dpi=300)
plt.plot(x,y,label='y=cos(x)')
plt.plot(x,y2,label='y=sin(x)')

ax.set_xticks(a,
           r,
           fontsize=20
           )
ax.tick_params(axis='x', labelrotation=0)
ax = plt.gca()
plt.legend(frameon=False,loc='best')
ax.spines['left'].set_position(('data',0)) 
ax.spines['bottom'].set_position(('data',0)) 
ax.spines['top'].set_color(None) 
ax.spines['right'].set_color(None) 
ax.tick_params(axis='x',labelsize=14)
```
