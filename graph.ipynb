```python
import numpy as np
import matplotlib.pyplot as plt
import fractions
```

```python
x = np.linspace(-2*np.pi,np.pi*2,1000)
y = np.cos(x)
y2 = np.sin(x)
fig, ax=plt.subplots(figsize=(10,6),dpi=300)
plt.plot(x,y,label='y=cos(x)')
plt.plot(x,y2,label='y=sin(x)')

ax.set_xticks([i for i in np.arange(-2*np.pi,2*np.pi+0.1,np.pi/4)],
           [f'{fractions.Fraction(round(i/np.pi,2))}$\pi$' for i in np.arange(-2*np.pi,2*np.pi+0.1,np.pi/4)],
           )
ax.tick_params(axis='x', labelrotation=45)
ax = plt.gca()
plt.legend(frameon=False,loc='best')
ax.spines['left'].set_position(('data',0)) 
ax.spines['bottom'].set_position(('data',0)) 
ax.spines['top'].set_color(None) 
ax.spines['right'].set_color(None) 
```
